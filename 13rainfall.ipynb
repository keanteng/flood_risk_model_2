```python
import pandas as pd
```

```python
# csv merge
file = [
    'cleaned/2020_all_states.xlsx',
    'cleaned/2021_all_states.xlsx'
]
```

```python
# merging two csv files
df = pd.concat(
    map(pd.read_excel, file), ignore_index=True)
df.head()
```

```python
df['nilai_hujan_max'] = df['nilai_hujan_max'].fillna(method = 'bfill')
df['tempoh_hujan'] = df['tempoh_hujan'].fillna(method = 'bfill')
df['tempoh_ulang'] = df['tempoh_ulang'].fillna(method = 'bfill')
df['kawasan_banjir'] = df['kawasan_banjir'].fillna(value = "Unkonwn Location")
df["kedalaman_banjir_max"] = df["kedalaman_banjir_max"].fillna(method = "bfill")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(how = "any", axis = 0)
df.head()
```

```python
df.tail()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
len(df)
```

- we will export this data and sent to excel in order for the length to match with the regression dataset, since some values are deleted from the regression dataset

```python
df.to_csv("rainfall.csv", index=False)
```
